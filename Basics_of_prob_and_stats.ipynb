# Basics of R programming

(1) Create a vector c = [5,10,15,20,25,30] and write a program which returns the maximum and minimum of this vector.

(2) Write a program in R to find factorial of a number by taking input from user. Please print error message if the input number is negative.

(3) Write a program to write first n terms of a Fibonacci sequence. You may take n as an input from the user.

(4) Write an R program to make a simple calculator which can add, subtract, multiply and divide.

(5) Explore plot, pie, barplot etc. (the plotting options) which are built-in functions in R.

```r
# Problem - 1

vec<-c(5,10,15,20,25,30)
cat("Max_val = ",max(vec),"\n")
cat("Min_val = ",min(vec),"\n")
```

```r
# Problem - 2

fact_fun <- function(n){
  if(n<0){
    return("Error:Invalid!")
  }else{
    fact <- 1
    for(i in 1:n){
      fact <- fact*i
    }
    return(fact)
  }
}
num <- 6
result <- fact_fun(num)
cat("Factorial of ",num," is ",result,"\n")
```

```r
# Problem - 3

fibonacci <- function(n){
  fsq <- c()
  fsq[1] <- 0
  if(n>1){
    fsq[2] <- 1
  }
  for(i in (3:n)){
    fsq[i]<-fsq[i-1]+fsq[i-2]
  }
  return(fsq)
}

n_terms <- 8
fseq <- fibonacci(n_terms)
cat("fseq : ",fseq)
```

```r
# Problem 4

calculator <- function(a,b,opr){
  if(opr=='+'){
    return(a+b)
  }else if(opr=='-'){
    return(a-b)
  }else if(opr=='*'){
    return(a*b)
  }else if(opr=='/'){
    return(a/b)
  }else{
    return("Error!")
  }
}
a<-15
b<-2
opr<-'/'
res <- calculator(a,b,opr)
cat("\nResult : ",res)
```

```r
# Problem 5

x<- c(1,3,5,8,11)
y<- c(3,6,7,15,19)
plot(x,y,col="blue",type="b",main="Basic Plot",xlab="X axis",ylab="Y axis")
```

```r
data<- c(25,35,40)
label<- c("A","B","C")
pie(data,labels=label,col=c("red","yellow","green"),main="Pie Chart")
```

```r
bar_data<- c(5,10,15,20,25)
barplot(bar_data,main="Bar Plot",col="black",names.arg=c("A","B","C","D","E"))
```
